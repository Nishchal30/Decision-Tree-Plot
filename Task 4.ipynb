# Decision Tree

```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/Asus/Downloads/iris.csv')
```

```python
data.head()
```

```python
data.drop('Id', axis = 1, inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=1)
```

```python
x_train.shape, x_test.shape, y_test.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
model = tree.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('accuracy_score is{}:'.format(accuracy_score(y_test, y_pred)))
```

```python
data.feature_names = data.columns
```

```python
data.feature_names
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(80,55))
plot_tree(model, filled=True, rounded=True, feature_names=data.feature_names)
```
